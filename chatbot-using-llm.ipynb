```python
import torch
from transformers import AutoTokenizer, pipeline
```

```python
model = "tiiuae/falcon-7b-instruct"
```

```python
tokenizer= AutoTokenizer.from_pretrained(model)
```

```python
pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype = torch.bfloat16,
    device_map = "auto",
    trust_remote_code=True
)
```

```python
prompt = "What is operating system?"
```

```python
newline_token = tokenizer.encode("\n")[0]

my_name = "Ovi"
bot_name = "Bot"
dialog = [
    f"The following is a conversation between {my_name} (a user) and {bot_name} (an AI assistant).",
    f"{my_name}: Hi",
    f"{bot_name}: Hello! How can I help you today?"
]

while True:
    user_input = input("Ovi: ")
    #print(f"{my_name}: {user_input}")
    dialog.append(f"{my_name}: {user_input}")
    
    # Format prompt
    prompt = "\n".join(dialog) + f"\n{bot_name}:"

    sequences = pipeline(
        prompt,
        max_length=len(tokenizer.encode(prompt)) + 50,
        do_sample=True,
        top_k=10,
        temperature=0.7,
        num_return_sequences=1,
        return_full_text=False,
        eos_token_id=newline_token,
        pad_token_id=tokenizer.eos_token_id,
    )

    bot_reply = sequences[0]["generated_text"].strip()

    # Clean any hallucinated speaker tags
    for tag in [f"{my_name}:", f"{bot_name}:"]:
        if tag in bot_reply:
            bot_reply = bot_reply.split(tag)[0].strip()

    print(f"{bot_name}: {bot_reply}")  # ðŸ‘ˆ Show bot reply with speaker tag
    dialog.append(f"{bot_name}: {bot_reply}")
```
